```python
# Clean the 'CC/Battery Capacity' column by removing non-numeric characters and converting to numeric
df['CC/Battery Capacity'] = df['CC/Battery Capacity'].astype(str).str.replace('[^\d\.]', '', regex=True)
df['CC/Battery Capacity'] = pd.to_numeric(df['CC/Battery Capacity'], errors='coerce')

# Calculate the mean of 'CC/Battery Capacity'
mean_cc = df['CC/Battery Capacity'].mean()

# Fill missing values with the mean
df['CC/Battery Capacity'].fillna(mean_cc, inplace=True)

# Verify that missing values have been filled
display(df.isnull().sum())
```

```python
# Clean the 'Torque' column by removing non-numeric characters and converting to numeric
df['Torque'] = df['Torque'].astype(str).str.replace('[^\d\.]', '', regex=True)
df['Torque'] = pd.to_numeric(df['Torque'], errors='coerce')

# Calculate the mean of 'Torque'
mean_torque = df['Torque'].mean()

# Fill missing values with the mean
df['Torque'].fillna(mean_torque, inplace=True)

# Verify that missing values have been filled
display(df.isnull().sum())
```

```python
# Function to parse price ranges and clean prices
def parse_price(value):
    if pd.isnull(value):
        return np.nan
    value = str(value).replace('$', '').replace(',', '').strip()
    if '-' in value:
        parts = [p.strip() for p in value.split('-')]
        if len(parts) == 2:
            try:
                low, high = float(parts[0]), float(parts[1])
                return (low + high) / 2
            except:
                return np.nan
    try:
        return float(value)
    except:
        return np.nan
```

```python
# Apply parsing to Cars Prices to create Cars Prices Clean
df['Cars Prices Clean'] = df['Cars Prices'].apply(parse_price)
```

```python
# Impute missing prices with mean
mean_price = df['Cars Prices Clean'].mean()
df['Cars Prices Clean'].fillna(mean_price, inplace=True)
```

```python
# Function to parse other numerical columns (from previous code, for consistency)
def parse_range(value):
    if pd.isnull(value):
        return np.nan
    value = str(value).replace(',', '').replace(' hp', '').replace(' Nm', '').replace(' km/h', '').replace(' sec', '').replace(' cc', '').replace(' kWh', '').strip()
    if '-' in value:
        parts = [p.strip() for p in value.split('-')]
        if len(parts) == 2:
            try:
                low, high = float(parts[0]), float(parts[1])
                return (low + high) / 2
            except:
                return np.nan
    elif '/' in value:
        parts = [p.strip() for p in value.split('/')]
        if len(parts) == 2:
            try:
                low, high = float(parts[0]), float(parts[1])
                return (low + high) / 2
            except:
                return np.nan
    elif '+' in value:
        parts = [p.strip() for p in value.split('+')]
        try:
            return sum(float(p) for p in parts)
        except:
            return np.nan
    try:
        return float(value)
    except:
        return np.nan
```

```python
# Apply parsing to other numerical columns
num_cols_to_clean = ['CC/Battery Capacity', 'HorsePower', 'Total Speed', 'Performance(0 - 100 )KM/H', 'Torque', 'Seats']
for col in num_cols_to_clean:
    df[col] = df[col].apply(parse_range)
```

```python
# Impute missing values with means
for col in num_cols_to_clean:
    mean_val = df[col].mean()
    df[col].fillna(mean_val, inplace=True)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('Cars Datasets 2025.csv', encoding='latin1')

#Convert numeric columns
def to_numeric_clean(s):
    s = s.astype(str).str.replace('[^\d\.]', '', regex=True)
    s = pd.to_numeric(s, errors='coerce')
    return s

df['speed_num'] = to_numeric_clean(df['Total Speed'])
df['price_num'] = to_numeric_clean(df['Cars Prices'])
df['torque_num'] = to_numeric_clean(df['Torque'])
df['hp_num'] = to_numeric_clean(df['HorsePower'])
df['cc_num'] = to_numeric_clean(df['CC/Battery Capacity'])
df['accel_num'] = to_numeric_clean(df['Performance(0 - 100 )KM/H'])
```

```python
# Create new features

df['price_per_speed'] = df['price_num'] / df['Total Speed'].replace({0: np.nan})
df['price_per_torque'] = df['price_num'] / df['Torque'].replace({0: np.nan})
df['price_per_hp']   = df['price_num'] / df['hp_num'].replace({0: np.nan})
df['price_per_cc']   = df['price_num'] / df['cc_num'].replace({0: np.nan})
df['hp_per_cc']      = df['hp_num'] / df['cc_num'].replace({0: np.nan})
df['accel_score']    = 1 / df['accel_num'].replace({0: np.nan})
df['power_to_price'] = df['hp_num'] / df['price_num'].replace({0: np.nan})
```

```python
display(df.head())
```

```python
# Check skewness and apply log transform

print("Skewness before log transformation:")
print(df.select_dtypes(include=[np.number]).skew())

num_cols = ['speed_num','torque_num','price_num','hp_num','cc_num','accel_num','price_per_hp','price_per_cc','hp_per_cc','power_to_price']

skewness = df[num_cols].skew().sort_values(ascending=False)
print("Skewness:\n", skewness)

skewed = skewness[skewness.abs() > 0.75].index.tolist()
for col in skewed:
    if (df[col].dropna() >= 0).mean() > 0.9:  # ensure non-negative
        df[f"{col}_log"] = np.log1p(df[col])
```
